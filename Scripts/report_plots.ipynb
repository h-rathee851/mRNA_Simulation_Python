```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = open("final_data/final_data.dat",'r')
```

```python
lines = file.readlines()
```

```python
L_ = []
alpha_ = []
w_ = []
gene_name_ = []
```

```python
for line in lines:
    
    data = line.split(',')
    
    if 'N/A' not in data[3]:
        
        gene_name_.append(data[0])
        L_.append(int(data[1]))
        alpha_.append(round(float(data[3]),4))
        
        
    
    
    
```

```python
file.close()
```

```python
for gene in gene_name_:
    
    file_name = '../elongation-rates/'+gene+'_rates.dat'
    file = open(file_name,'r')
    lines = file.readlines()
    
    for line in lines:
        
        data = line.split(' ',1)
        
        w_.append(float(data[1]))
        
    file.close()
```

```python
plt.hist(w_,bins = 17,color='red')
plt.xlabel(r"$\omega_{i}(1/s)$")
plt.ylabel("Number of sites")
plt.title("Histogram of Distribution of hopping rates for Baker's Yeast")
plt.show()
```

```python
plt.hist(alpha_, bins = 50,color='turquoise')
plt.xlabel(r"Value of $\alpha(1/s)$")
plt.ylabel("Number of Genes")
plt.title(r'Histogram of Distribution of $\alpha$ for '+str(len(alpha_))+' genes of baker\'s yeast')
plt.show()
```

```python
fig = plt.figure()
ax = plt.gca()
ax.scatter(L_ ,alpha_ , c='blue', alpha=0.15, edgecolors='none')
ax.set_yscale('log')
ax.set_ylim(0.007,1)
ax.set_xlim(80,1500)
ax.set_xlabel("Lattice Size")
ax.set_ylabel(r"$\alpha(1/s)$")
ax.set_title(r"Log-Log plot of L vs $\alpha$")
ax.set_xscale('log')
plt.show()
```
